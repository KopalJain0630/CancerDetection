```python
import pandas as pd
import numpy as np
import seaborn as sns
import shap
```

```python
df = pd.read_csv('data.csv')

#view the data
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(axis = 1, inplace = True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1] = labelencoder.fit_transform(df.iloc[:,1].values)
```

```python
df
```

```python
X = df.iloc[:,2:].values 
y = df.iloc[:,1].values.astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predictions)
sns.heatmap(cm,annot = True)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
explainer = shap.Explainer(classifier, X_train)

# Calculate SHAP values for the test set
shap_values = explainer(X_test)
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")  # Feature importance as bar plot
shap.summary_plot(shap_values, X_test)
```

```python
sample_index = 0  # Select a sample from X_test
shap.waterfall_plot(shap_values[sample_index])
```

```python
feature_name = "Feature 27"  # Adjust this to any feature in your dataset
shap.dependence_plot(feature_name, shap_values.values, X_test)
```
