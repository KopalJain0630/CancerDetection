```python
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN(random_state=42)
X, y = adasyn.fit_resample(X, y)
```

```python
len(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

```python
explainer = shap.Explainer(lr_model, X_train)

# Calculate SHAP values for the test set
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")  # Feature importance as bar plot
shap.summary_plot(shap_values, X_test)
```

```python
sample_index = 0  # Select a sample from X_test
shap.waterfall_plot(shap_values[sample_index])
feature_name = "CHRONIC DISEASE"  # Adjust this to any feature in your dataset
shap.dependence_plot(feature_name, shap_values.values, X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)  
dt_model.fit(X_train, y_train)
```

```python
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

```python
explainer = shap.TreeExplainer(dt_model)

# Calculate SHAP values for the test set
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")  # Feature importance as bar plot
shap.summary_plot(shap_values, X_test)    
sample_index = 0  # Select an index of a sample in X_test
shap.waterfall_plot(shap.Explanation(values=shap_values[1][sample_index], 
                                     base_values=explainer.expected_value[1], 
                                     data=X_test.iloc[sample_index]))
feature_name = "CHRONIC DISEASE"  # Adjust this to any feature in your dataset
shap.dependence_plot(feature_name, shap_values, X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
```

```python
mnb_model_scores = cross_val_score(mnb_model,X, y, cv=kf)

# Support Vector Classifier model
svc_model_scores = cross_val_score(svc_model,X, y, cv=kf)

# Random forest model
rf_model_scores = cross_val_score(rf_model,X, y, cv=kf)

# XGBoost model
xgb_model_scores = cross_val_score(xgb_model,X, y, cv=kf)

# Multi-layer perceptron model
mlp_model_scores = cross_val_score(mlp_model,X, y, cv=kf)

# Gradient boost model
gb_model_scores = cross_val_score(gb_model,X, y, cv=kf)
```
